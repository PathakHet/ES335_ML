Answer 1:

After multiplication of a matrix A(n, k) and B(k, m), we know that we will get a matrix C (say), of dimension (n, m). Now, to calculate each element C[i][j], we need the dot product of the ith row of A and jth column of B. Both the ith row of A and the jth column of B contains k elements. This means that there will be k corresponding multiplications and (k-1) additions to sum the k elements thus obtained.

Therefore, for forming the complete matrix C, of nm elements, we will require: nmk multiplications; nm*(k-1) additions.

```python
# Answer 2
import numpy as np
import time

mat1 = [[1, 2, 3, 4], [4, 5, 6, 7], [7, 8, 9, 10]]
mat2 = [[3, 1, 0, 4], [2, 1, 1, 3], [4, 2, 3, 2], [4, 5, 2, 3]]
result = np.zeros((len(mat1), len(mat2[0])), dtype=int)

# multiplying without numpy
time1 = time.time()
for i in range(len(mat1)): # rows of mat1
    for j in range(len(mat2[0])): # columns of mat2
        for k in range(len(mat2)): # rows of mat2
            result[i][j] += mat1[i][k] * mat2[k][j]

print(result)
time2 = time.time()


# multiplying using numpy
result1 = np.dot(mat1, mat2)
print(result1)
time3 = time.time()

print("Time taken for multiplication without numpy:", time2 - time1)
print("Time taken for multiplication using numpy:", time3 - time2)
if (time2-time1) < (time3-time2):
    print("Numpy multiplied it slower")
else:
    print("Numpy multiplied it faster")
```

When you perform matrix multiplication using a list of lists in Python, the operations are carried out using Python's built-in operations, which are not as optimized for numerical computations as the operations in NumPy, which utilizes broadcasting for smooth operations. When you perform matrix multiplication using NumPy, the underlying operations are executed using highly efficient and optimized code, which takes advantage of hardware-specific optimizations and parallel processing capabilities.

Using the described method will require about n/2 (minor difference based on odd,even number of elements in the array) passes of the array, and one pass is order n, thus finding the median will be O(n^2/2)

```python
# Using given method in the question

np.random.seed(36)

array = np.random.rand(5000)

n = len(array)

start_time = time.time()
for q in range((len(array) // 2) + 1):
    max_index = 0
    for i in range(n):
        if array[i] > array[max_index]:
            max_index = i
    array[max_index], array[n - 1] = array[n - 1], array[max_index]
    n -= 1
median = array[n]
end_time = time.time()
time_taken = end_time - start_time

print("Median:", median)
print("Time Taken:", time_taken)
```

A better method would be to sort the array in O(nlog(n)) and find the middle element.

```python
array2 = np.random.rand(5000)


start_time = time.time()

array2.sort()
median = array2[len(array2) // 2]

end_time = time.time()
time_taken2 = end_time - start_time
print("Median:", median)
print("Time Taken:", time_taken2)
```

Now using numpy median.

```python
array3 = np.random.rand(5000)


start_time = time.time()

median = np.median(array3)

end_time = time.time()
time_taken3 = end_time - start_time
print("Median:", median)
print("Time Taken:", time_taken3)
```

We see that the method given in the question takes the longest time. Then sorting using nlog(n) and numpy.median is the fastest for long inputs.

$f(x, y) = x^2*y + y^3*sin(x)$

The gradient of f with respect to x is:

$df/dx = 2x*y + y^3*cos(x)$

The gradient of f with respect to y  is:

$df/dy = x^2 + 3y^2*sin(x)$

Answer 5

```python
import jax
import jax.numpy as jnp

def f(x, y):
    return x**2*y + y**3*jnp.sin(x)


grad_f = jax.grad(f, argnums=(0, 1))

# Evaluate the gradients at random values of x and y
x1 = 4.0
y1 = 6.0
grad_x, grad_y = grad_f(x1, y1)


analytical_grad_x = 2*x1*y1 + y1**3*jnp.cos(x1)
analytical_grad_y = x1**2 + 3*y1**2*jnp.sin(x1)
print("Analytical Gradient:")
print("df/dx =", analytical_grad_x)
print("df/dy =", analytical_grad_y)

print("\nJAX Computed Gradient:")
print("df/dx =", grad_x)
print("df/dy =", grad_y)
```

```python
import sympy as sp

x, y = sp.symbols('x y')

f = x**2 * y + y**3 * sp.sin(x)

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Analytical Gradient with respect to x:", df_dx)
print("Analytical Gradient with respect to y:", df_dy)
```

```python
database = []


class Year:
    def __init__(self, year, branches=None):
        self.year = year
        self.branches = branches or []

    def add_branch(self, branch):
        self.branches.append(branch)


class Branch:
    def __init__(self, name, students=None):
        self.name = name
        self.students = students or []


class Student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

student1 = Student('N', 1, {'Maths': 100, 'English': 70})
student2 = Student('M', 2, {'Maths': 90, 'English': 80})

branch1 = Branch('Branch 1', [student1, student2])
branch2 = Branch('Branch 2')

year2022 = Year(2022, [branch1, branch2])
year2023 = Year(2023)

database.append(year2022)
database.append(year2023)

print("English marks of student M in 2022, Branch 1:", database[0].branches[0].students[1].marks['English'])
```
